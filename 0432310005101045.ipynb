```python
# 1. Uniform Random Numbers:
# Use np.random.rand() to generate a 1×10 vector of random numbers between 0 and 1.

#Plot a histogram of the 1×10 vector to visualize the distribution.
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

random_numbers = np.random.rand(10)

print("Random Numbers:")
print(random_numbers)
plt.hist(random_numbers, bins=5, edgecolor='black', color='red')

plt.title("Histogram of 10 Uniform Random Numbers (0 to 1)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
#2.  Normally Distributed Random Numbers:

#Generate 500 random numbers from a standard normal distribution (mean=0, standard deviation=1) using np.random.normal.
#Plot a histogram of these numbers to visualize the normal distribution.

import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

NormalValues = np.random.normal(0, 1, 500)

# Create a histogram
plt.hist(NormalValues, bins=30, edgecolor='black')
plt.title('Histogram of Normal Distribution (mean=0, std=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
